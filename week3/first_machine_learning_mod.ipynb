```python
import pandas as pd
metro = pd.read_csv('Standard Metropolitan Areas Data - train_data - data.csv')
```

```python
metro.columns
```

```python
# Drops rows with missing data
metro = metro.dropna(axis = 0)
```

```python
metro.head()
```

```python
# Sets y to income column and x to all over columns
y = metro.income
x = metro.drop('income', axis = 1)
```

```python
x.describe()
```

```python
x.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x, y)
```

```python
print("Making predictions for the following 5 incomes:")
print(x.head())
print("The predictions are")
print(model.predict(x.head()))
```
