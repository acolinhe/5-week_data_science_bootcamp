```python
import numpy as np
import pandas as pd
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/dphi-official/First_ML_Model/master/titanic.csv')
```

```python
titanic.head()
```

```python
titanic['Sex'] = titanic['Sex'].apply(lambda x: 1 if x == 'male' else 0)
```

```python
titanic.isnull().sum()
```

```python
titanic['Age'] = titanic['Age'].fillna(np.mean(titanic['Age']))
```

```python
titanic.Embarked.value_counts()
```

```python
titanic.Embarked.fillna(value='S', axis = 0, inplace = True)
```

```python
titanic.isnull().sum()
```

```python
# Omit irrelavent columns
titanic = titanic[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
```

```python
X = titanic.drop('Survived', axis = 1)
y = titanic['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
# The acurracy function lets us know how well our model performs
from sklearn.metrics import accuracy_score
print('Training accuracy...', accuracy_score(y_train, model.predict(x_train)))
print('Test accuracy', accuracy_score(y_test, model.predict(x_test)))
```

```python
from sklearn.tree import export_graphviz
import graphviz
```

```python
dot_data = export_graphviz(model, out_file = None,
feature_names = x_test.columns,
class_names = ['0', '1'],
filled = True, rounded = True,
special_characters = True)
```
